```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)

jeopardy_data = pd.read_csv('jeopardy.csv')
print(jeopardy_data)
```

```python
##Inspect the columns and data types; notice leading space on most of the columns
print(jeopardy_data.columns)
print(jeopardy_data.dtypes)
```

```python
##Update column names to remove space
jeopardy_data.columns = ['Show Number', 'Air Date', 'Round', 'Category', 'Value',
       'Question', 'Answer']
print(jeopardy_data.columns)
```

```python
##Write a function that filters dataset for questions containing all of the words in a list of words. Utilize lambdas.

def is_in_question(data, words):
    filtered = lambda w: all(word.lower() in w.lower() for word in words)
    return data.loc[data["Question"].apply(filtered)]
    
sky_and_clouds = is_in_question(jeopardy_data, ['Sky', 'Clouds']) 
print(sky_and_clouds)
print(is_in_question(jeopardy_data, ['space', 'star']))
```

```python
##Convert the "Value" column to float.  Remove $ and ,  values
jeopardy_data['Value'] = jeopardy_data['Value'].apply(lambda v:float(v[1:].replace(',','')) if v != "None" else 0)
print(jeopardy_data.Value.head(10))
```

```python
##Print average value of questions that contain certain words above
space_and_star = is_in_question(jeopardy_data, ['space', 'star'])
print(space_and_star.Value.mean())
```

```python
#Write a function that returns count of unique answers to all of the questions in a dataset.

def unique_answers(data):
    return data['Answer'].value_counts()

print(unique_answers(space_and_star))
```
